```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

base_model = "tiiuae/falcon-7b-instruct"

tokenizer = AutoTokenizer.from_pretrained(base_model, device_map="auto")
model = AutoModelForCausalLM.from_pretrained(base_model, device_map="auto", load_in_4bit=True)

model = model.eval()
```

```python
import sys
sys.path.append('../../benchmarks/')

from sentiment140 import test_sentiment140

dataset = test_sentiment140(model, tokenizer)
```

```python
dataset.to_csv('../../results/general/falcon-7b-instruct.csv')
```
