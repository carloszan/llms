```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

base_model = "tiiuae/falcon-7b-instruct"

tokenizer = AutoTokenizer.from_pretrained(base_model, device_map="auto")
model = AutoModelForCausalLM.from_pretrained(base_model, device_map="auto", load_in_8bit=True, torch_dtype=torch.float16)

model = model.eval()
```

```python
import sys
sys.path.append('../benchmarks/')

from fiqa import test_fiqa

dataset = test_fiqa(model, tokenizer)
```

```python
dataset.to_csv('../results/falcon-7b-instruct.csv')
```
